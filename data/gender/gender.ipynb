```python
import pandas as pd
```

```python
gender_df = pd.read_csv('gender.csv')
gender_df.update(gender_df['gender'].mask(gender_df['gender'] == 'w', 'f'))
gender_df
```

```python
speakers = pd.read_csv("../rolle.tsv", sep="\t")
additional_speakers = pd.read_csv("../additional_roles.csv", sep=",")
speakers = pd.concat([speakers, additional_speakers], ignore_index=True).drop_duplicates(subset=['name'], keep='first')
print(len(speakers[speakers.duplicated(subset=['name'], keep=False)]), 'duplicates found')
speakers
```

```python
w_identifier = [
    'mrs ',
    'miss ',
    'ms ',
    'mrs. ',
    'miss. ',
    'ms. ',
    'tante ',
    'lady'
]
w = []

m_identifier = [
    'mister ',
    'mr '
    'mister. ',
    'mr. ',
    'onkel '
]
m = []

for i, row in speakers.iterrows():
    if row['rolleID'] == -1:
        continue  # skip UNKNOWNs

    for ident in w_identifier:
        if row['name'].lower().startswith(ident):
            w.append((row['rolleID'], row['name']))
    
    for ident in m_identifier:
        if row['name'].lower().startswith(ident):
            m.append((row['rolleID'], row['name']))
print(f'found {len(w)} w')
print(f'found {len(m)} m')
```

```python
m
```

```python
m_df = pd.DataFrame({
    'rolleID': [entry[0] for entry in m],
    'gender': ['m' for _ in range(len(m))]
})
m_df
```

```python
w_df = pd.DataFrame({
    'rolleID': [entry[0] for entry in w],
    'gender': ['w' for _ in range(len(w))]
})
w_df
```

```python
new_gender_df = pd.concat([gender_df, m_df, w_df], ignore_index=True).drop_duplicates(subset=['rolleID'], keep='first')
new_gender_df = new_gender_df.sort_values(by='rolleID')
new_gender_df
```

```python
new_gender_df.to_csv('gender.csv',index=False)
```
