```python
import pandas as pd
```

# CSV WITH SPEAKER NAMES -> gender.csv
- this part of the script imports 'speakers_with_gender.csv' and exports it (without the names) as 'gender.csv'
- you can edit the 'speakers_with_gender.csv' file inside Excel and this part of the script will process it

```python
gender_df = pd.read_csv('speakers_with_gender.csv')
gender_df = gender_df[~gender_df['gender'].isna()]
gender_df = gender_df.drop_duplicates(subset=['rolleID'], keep='first')
gender_df = gender_df.drop(columns=['name'])
print(len(gender_df[gender_df['gender'] == 'f']), 'female')
print(len(gender_df[gender_df['gender'] == 'm']), 'male')
gender_df
```

```python
gender_df.to_csv('gender.csv',index=False)
```

# CSV WITHOUT SPEAKER NAMES -> speakers_with_gender.csv
- this part of the script imports 'gender.csv' and exports it (with the names) as 'speakers_with_gender.csv'
- you can then edit the 'speakers_with_gender.csv' file inside Excel and the previous part of the script will process it

```python
gender_df = pd.read_csv('gender.csv')
gender_df = gender_df[~gender_df['gender'].isna()]
gender_df = gender_df.drop_duplicates(subset=['rolleID'], keep='first')
print(len(gender_df[gender_df['gender'] == 'f']), 'female')
print(len(gender_df[gender_df['gender'] == 'm']), 'male')
gender_df
```

```python
# LOAD SPEAKERS
speakers = pd.read_csv("../rolle.tsv", sep="\t")
speakers
```

```python
speakers_with_gender = speakers.merge(gender_df, on='rolleID', how='left')
speakers_with_gender
```

```python
speakers_with_gender.to_csv('speakers_with_gender.csv',index=False)
```
